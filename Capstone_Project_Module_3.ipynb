###Menghitung Jumlah Outlier menggunakan Inter Quartile Range (IQR) * 1.5

```python
list_jumlah_outliers = []
for setiap_kolom in numerik.columns:
    q3 = numerik.describe()[setiap_kolom]['75%']
    q1 = numerik.describe()[setiap_kolom]['25%']
    iqr = q3 - q1
    whis = 1.5 * iqr
    upper_bound = q3 + whis
    lower_bound = q1 - whis

    jumlah_outliers = len(numerik[(numerik[setiap_kolom] > upper_bound) | (numerik[setiap_kolom] < lower_bound)])
    list_jumlah_outliers.append(jumlah_outliers)

pd.DataFrame({
                'kolom': numerik.columns,
                'jumlah outliers': list_jumlah_outliers

})
```

###Multikolineartias dengan VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
# buat function
def calc_vif(x):
    # dataframe kosong
    vif = pd.DataFrame()
    vif['variables'] = x.columns
    # menghitung vig untuk setiap kolom nya
    vif['VIF'] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
    return vif
```

```python
calc_vif(numerik.drop(columns = 'Price')).sort_values(by = 'VIF', ascending = False)
```

Fungsi untuk cek outlier:

```python
def check_outlier(data):
  # Hitung Q1 dan Q3
  Q1 = data.quantile(0.25)
  Q3 = data.quantile(0.75)

  # Hitung IQR
  IQR = Q3 - Q1
  print(f'IQR: {iqr}')

  # Tentukan batas atas dan bawah
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  print(f"lower_bound: {lower_bound}")
  print(f"upper_bound: {upper_bound}")
```

dalam proses ini kita akan buat Column Transformer yang berisi One-Hot Encoder, Binary Encoder, Min Max Scaler serta Robust Scaler dengan alasan:
- One-Hot Encoder digunakan pada kolom Gear_Type, Origin, dan Option karena memiliki Unique Values dibawah 5.
- Binary Encoder digunakan pada kolom Type, Region, Make karena memiliki Unique Values 5 atau lebih.
- Min Max Scaler digunakan pada data  Year, dan Engine_Size karena tidak seperti StandardScaler, yang mengasumsikan data terdistribusi normal (berpusat di sekitar rata-rata 0), MinMaxScaler tidak membuat asumsi tentang distribusi data yang mendasarinya. Ini sangat berguna jika fitur Year, dan Engine_Size yang tidak terdistribusi normal atau memiliki outlier serta outlier pada data bukan merupakan outlier ekstrem, karena ia hanya menskalakan ulang nilai tanpa memengaruhi distribusinya.
- Robust Scaler digunakan pada data Mileage karena data berisi outlier ekstrem.

#Penentuan Evaluation Metrics

Kita akan menggunakan MAE sebagai metrics dalam menentukan model yang akan dipilih dengan alasan sebagai berikut:
- MAE mengukur rata-rata kesalahan dalam satuan yang sama dengan variable target 'Price'. Dalam hal ini, kesalahan rata-rata dihitung dalam satuan Riyals. Ini memberikan interpretasi yang mudah dan jelas karena menunjukkan secara langsung berapa besar rata-rata kesalahan prediksi model dalam memprediksi harga mobil.
- MAE tidak terlalu sensitif terhadap outliers dibandingkan metric seperti RMSE. Ini berarti bahwa jika ada beberapa harga mobil yang sangat ekstrem (mobil mewah dengan harga sangat tinggi), MAE tidak akan memberikan penalti besar terhadap kesalahan pada outliers tersebut.
- MAE memberikan penilaian yang adil terhadap performa model di seluruh rentang harga mobil.
- Karena variable target 'Price' memiliki rentang yang cukup luas, seperti pada dataset yang digunakan, MAE bisa digunakan untuk memberikan ukuran yang relatif netral terhadap keseluruhan performa model.

Lalu kita juga akan menggunakan MAPE dan R2 sebagai faktor pendukung MAE dan RMSE sebagai perbandingan terhadap MAE.

##Rule Based Model (Non ML)
Kali ini kita akan buat Rule Based Model non ML untuk nantinya akan dijadikan perbandingan dengan model ML terbaik dalam menilai error dari model.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
def rule_based(car):
    price = 0

    # Rule 1: Sesuaikan harga berdasarkan tahun mobil
    if car['Year'] >= 2020:
        price += 120000
    elif 2015 <= car['Year'] < 2020:
        price += 90000
    elif 2010 <= car['Year'] < 2015:
        price += 60000
    else:
        price += 30000

    # Rule 2: Sesuaikan harga berdasarkan jarak tempuh
    if car['Mileage'] < 50000:
        price += 20000
    elif 50000 <= car['Mileage'] < 100000:
        price += 10000
    else:
        price -= 5000

    # Rule 3: Sesuaikan harga berdasarkan ukuran mesin
    if car['Engine_Size'] > 3.0:
        price += 15000
    elif 2.0 <= car['Engine_Size'] <= 3.0:
        price += 10000
    else:
        price += 5000

    # Rule 4: Sesuaikan harga berdasarkan brand/produsen mobil
    if car['Make'] == 'Toyota':
        price += 20000
    elif car['Make'] == 'Mercedes':
        price += 40000
    elif car['Make'] == 'BMW':
        price += 35000
    else:
        price += 10000  # Harga default untuk brand lain

    return price
```

```python
y_pred_train_rule_based = X_train.apply(rule_based, axis = 1)
score_train_rule_based = mean_absolute_error(y_train, y_pred_train_rule_based)

y_pred_test_rule_based = X_test.apply(rule_based, axis = 1)
score_test_rule_based = mean_absolute_error(y_test, y_pred_test_rule_based)
```

```python
df_rule_model = pd.DataFrame({
    'Model': ['Rule Based Model'],
    'Train MAE': [score_train_rule_based],
    'Test MAE': [score_test_rule_based]
})
df_rule_model
```

## Experiment 1: Based Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import VotingRegressor, StackingRegressor, RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
#base Model
lr = LinearRegression()
knr = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state=0)

# voting
voting = VotingRegressor([
    ('lr', lr),
    ('dt', dt),
    ('knr', knr)
    ])

# stacking
base_model = ([
    ('lr', lr),
    ('dt', dt),
    ('knr', knr)
])
meta_model = LinearRegression()
stacking = StackingRegressor(base_model,meta_model)

# bagging
rf = RandomForestRegressor(random_state=0)

# boosting
gb = GradientBoostingRegressor(random_state=0)
xgb = XGBRegressor(random_state=0)
```

###Menggunakan MAE

```python
# Import libraries yg dibutuhkan
from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error

# List untuk menampilkan hasil
train_score_list = []
train_score_std_list = []
test_score_list = []

# Mendefinisikan model yang akan digunakan
models = [lr, dt, knr, voting, stacking, rf, gb, xgb]

# KFold cross-validation setup
kf = KFold(n_splits=5, shuffle=True, random_state=0)

# Loop kepada setiap model
for model in models:
    pipeline = Pipeline([
        ('transformer', transformer),
        ('model', model)
    ])

    # Train model dan kalkulasi cross-validation scores (pada training set)
    cv_scores = cross_val_score(pipeline, X_train, y_train, cv=kf, scoring='neg_mean_absolute_error')

    # Kalkulasi mean and std untuk CV scores
    train_score_list.append(-cv_scores.mean())  # Karena default negative, kita kalikan dengan negative agar mudah membaca nilai MAE
    train_score_std_list.append(cv_scores.std())

    # Fit model pada semua training data
    pipeline.fit(X_train, y_train)

    # Membuat predict kepada test set
    y_pred_test = pipeline.predict(X_test)

    # Evaluasi model pada test set menggunakan MAE
    test_score = mean_absolute_error(y_test, y_pred_test)
    test_score_list.append(test_score)

# Menampilkan hasil
for i, model in enumerate(models):
    print(f"Model: {model}")
    print(f"Train Mean MAE: {train_score_list[i]:.4f}")
    print(f"Train MAE Std Dev: {train_score_std_list[i]:.4f}")
    print(f"Test MAE: {test_score_list[i]:.4f}\n")
```

##Bayesian optimization
 Alasan penggunaan Bayesian Optimization karena cocok untuk hyperparameter tuning terutama ketika evaluasi model mahal atau ketika ruang hyperparameter yang ingin dieksplorasi sangat besar.

```python
from skopt import BayesSearchCV
from xgboost import XGBRegressor
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler, RobustScaler
from category_encoders import BinaryEncoder
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
```

```python
#Transformer
transformer = ColumnTransformer([
    ('onehot', OneHotEncoder(drop = 'first'), ['Gear_Type', 'Origin', 'Options']),
    ('binary', BinaryEncoder(), ['Type', 'Region', 'Make']),
    ('minmax', MinMaxScaler(), ['Year', 'Engine_Size']),
    ('robust', RobustScaler(), ['Mileage'])
], remainder = 'passthrough')

# Pipeline
final_model = Pipeline([
    ('transformer', transformer),
    ('model', XGBRegressor(random_state=0))
])

# Parameter space for Bayesian Optimization
param_space = {
    'model__n_estimators': (50, 1000),           # Number of boosting rounds
    'model__max_depth': (3, 10),                 # Maximum depth of the trees
    'model__learning_rate': (0.01, 0.3, 'log-uniform'),  # Learning rate
    'model__subsample': (0.5, 1.0),              # Rasio subsampling dari contoh pelatihan
    'model__colsample_bytree': (0.5, 1.0),       # Rasio subsampling kolom untuk setiap pohon
    'model__gamma': (0, 1),                      # Pengurangan kerugian minimum diperlukan untuk pemisahan
    'model__reg_lambda': (0, 10),                # L2 regularization term
    'model__reg_alpha': (0, 10),                 # L1 regularization term
}

# Bayesian Optimization with Cross-Validation
opt = BayesSearchCV(
    final_model,
    param_space,
    n_iter=50,  # Jumlah iterasi untuk pencarian
    cv= KFold(5),       # 5-fold cross-validation
    scoring='neg_mean_absolute_error',  # Optimization metric (MAE)
    random_state=0,
    n_jobs=-1,  # Use all available cores
    verbose=3
)

# Fit the Bayesian search
opt.fit(X_train_without_outliers, y_train_without_outliers)
```

#Penjelasan Model

Model yang ditunjukkan dalam diagram diatas merupakan pipeline machine learning yang dirancang untuk memprediksi harga mobil bekas berdasarkan dataset Saudi Arabia Used Cars. Berikut penjelasan dari komponen-komponen utama dan prosesnya:

1. ColumnTransformer:
    
    Transformer ini melakukan serangkaian langkah preprocessing pada dataset:
    - OneHotEncoder (drop='first'): Variabel kategorikal seperti Gear_Type, Origin, dan Options diubah menjadi fitur biner. Opsi drop='first' mencegah collinearity dengan menghapus kategori pertama dari setiap variabel.
    - BinaryEncoder: Untuk variabel kategorikal dengan banyak nilai unik (Type, Region, Make), digunakan metode binary encoding yang mengonversi variabel tersebut ke dalam format numerik.
    - MinMaxScaler: Diterapkan pada fitur numerik seperti Year, Engine_Size, dan Mileage. Fitur-fitur ini diskalakan dalam rentang 0 hingga 1 agar bisa dibandingkan secara adil.
    - RobustScaler: Diterapkan pada Mileage untuk membuat model kurang sensitif terhadap outlier dengan melakukan scaling berdasarkan rentang interkuartil (IQR).
    - Remainder='passthrough': Kolom-kolom yang tidak disebutkan secara eksplisit dalam transformasi akan diteruskan tanpa perubahan.


2. Model - XGBoost Regressor (XGBRegressor):
    
    Setelah preprocessing, model ini menggunakan XGBoost Regressor untuk memprediksi harga mobil. XGBoost adalah metode ensemble yang kuat berbasis gradient boosting, dikenal karena efisiensi dan performanya dalam tugas regresi maupun klasifikasi.
    
    Beberapa hyperparameter kunci dalam XGBoost Regressor:

    - learning_rate=0.037: Mengontrol seberapa cepat model belajar. Nilai yang lebih rendah berarti pembelajaran yang lebih lambat namun biasanya menghasilkan performa yang lebih baik.
    - max_depth=8: Menentukan kedalaman maksimum setiap pohon dalam model. Pohon yang lebih dalam bisa menangkap pola yang lebih kompleks, tetapi juga berisiko overfitting.
    - n_estimators=768: Jumlah pohon (weak learners) yang digunakan dalam model. Semakin banyak pohon umumnya akan meningkatkan performa hingga batas tertentu.
    - colsample_bytree=0.5: Menentukan fraksi fitur yang digunakan saat membangun setiap pohon, memperkenalkan elemen randomisasi dan mengurangi overfitting.
    - gamma=1: Mengontrol kompleksitas model. Nilai yang lebih besar menghasilkan model yang lebih sederhana (dengan lebih sedikit pemisahan pada pohon).
    - Parameter lain seperti subsample, min_child_weight, dan max_leaves juga berperan dalam mengontrol overfitting dan memastikan generalisasi.


Ringkasan:

Pipeline ini dirancang untuk menangani data kategorikal dan numerik dengan efisien, mengubah dan menskalakan fitur secara tepat sebelum dimasukkan ke dalam XGBoost Regressor. Hyperparameter pada model XGBoost telah di-tuning (kemungkinan menggunakan Bayesian Optimization) untuk mencapai performa terbaik dalam memprediksi harga mobil di pasar mobil bekas Arab Saudi.

Model ini seharusnya robust terhadap berbagai jenis input data, mampu menangani outlier dengan baik (melalui RobustScaler), dan memanfaatkan fleksibilitas serta kekuatan XGBoost dalam menangani data terstruktur untuk tugas regresi.
